# Création d'un perceptron à la main
## Description

Un perceptron est un modèle simplifié de neurone artificiel utilisé en apprentissage automatique et en intelligence artificielle. 

Il est inspiré du fonctionnement des neurones dans le cerveau humain. Un perceptron prend plusieurs entrées, les pondère et les combine linéairement pour produire une sortie. Cette sortie est généralement passée à travers une fonction d'activation pour obtenir la réponse finale du perceptron. 

Il est souvent utilisé dans les tâches de classification binaire, où il apprend à séparer les données en deux catégories en ajustant les poids des entrées pendant l'entraînement

## 1ère Partie : Perceptron a 2 paramètres d'entrée

On importe le dataset make_blobs pour les essais

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.metrics import accuracy_score
```

Affichage du nuage de points

```python
X, y = make_blobs(n_samples = 100, n_features = 2, centers = 2, random_state = 0)
y = y.reshape((y.shape[0], 1))

plt.scatter(X[:, 0], X[:, 1], c = y, cmap = 'summer')

plt.show()
```

Création de la fonction d'initialisation des paramètres

```python
def init(x):
    return (np.random.rand(X.shape[1], 1), np.random.randn(1))
```

Création de la fonction d'activation, ici on utilise une sigmoïde $a(z) = \frac{1}{1+exp(-z)}$

```python
def model(x, w, b):
    z = x.dot(w) + b
    a_z = 1 / (1 + np.exp(-z))
    return a_z
```

Fonction de coût

```python
def log_loss(a, y):
    return 1/len(y) * np.sum(-(1-y) * np.log(1-a) - y * np.log(a))
```

Calcul des gradients

```python
def gradient(x, a, y):

    dw = (1 / len(y)) * np.dot(x.T, a - y)
    db = (1 / len(y)) * np.sum(a - y)
    
    return (dw, db)
```

```python
def update_neurone(w, b, dw, db, alpha):
    new_w = w - alpha * dw
    new_b = b - alpha * db

    return (new_w, new_b)
```

Modèle de prédiction, si la probabilité renvoyée par la sigmoïde est >= 0.5 on renvoie <code>True</code> sinon <code>False</code>

```python
def predict(x, w, b):
    a = model(x, w, b)
    return a >= 0.5
```

```python
from sklearn.metrics import accuracy_score
```

Le neurone en lui-même

```python
def artificial_neuron(X, y, learning_rate = 0.1, n_iter = 100):
    w, b = init(X)

    loss = []
    accuracy = []
    
    for i in range(n_iter):
        a = model(X, w, b)
        loss.append(log_loss(a, y))
        dw, db = gradient(X, a, y)
        w, b = update_neurone(w, b, dw, db, learning_rate)

        y_pred = predict(X, w, b)
        
        accuracy.append(accuracy_score(y_pred, y))

    y_pred = predict(X, w, b)
    accuracy_finale = accuracy_score(y, y_pred)
    print(f'Accuracy du modèle : {accuracy_finale}')

    min_loss = min(loss)
    
    print(f'Minimum atteint par la loss : {min_loss}')

    plt.figure(figsize = (12, 4))
    plt.subplot(1, 2, 1)
    plt.plot(loss, label = 'train loss')
    plt.legend()

    plt.subplot(1, 2, 2)
    plt.plot(accuracy, label = 'train accuracy')
    plt.legend()
    
    plt.show()

    return (w, b)
```

## Application du modèle

```python
w, b = artificial_neuron(X, y, learning_rate = 0.1, n_iter = 100)
```

On a une bonne convergence du modèle avec les hyperparamètres choisis

```python
# Pour le tracé de la frontière de décision
x1_line = np.array([-1, 4])
x2_line = - ( w[0] * x1_line + b )/w[1]
```

```python
new_plant = np.array([2, 1])
plt.scatter(X[:, 0], X[:, 1], c = y, cmap = 'summer')
plt.scatter(new_plant[0], new_plant[1], c = 'red')
plt.plot(x1_line, x2_line, linewidth = 2)

plt.show()
```

On voit qu'effectivement 14 points sont du mauvais côté de la frontière de décision.

#### Modifications des hyperparamètres

##### Modification du learning_rate (utilisé dans le calcul du gradient)

- $\alpha = 1$

```python
w, b = artificial_neuron(X, y, learning_rate = 1, n_iter = 100)
```

Convergence beaucoup plus rapide du modèle (lorsqu'on regarde la loss), on est proche du minimum après quelques itérations seulement. En revanche, l'accuracy ne semble pas "apprécier" cet apprentissage plus rapide.

- $\alpha = 5$

```python
w, b = artificial_neuron(X, y, learning_rate = 5, n_iter = 100)
```

On observe des "sauts" sur la loss sans aucun gain sur l'accuracy par rapport à $\alpha = 1$. Cela signifie que le modèle a du mal à converger vers un minimum local. Pour observer plus distinctement ce phénomène on augmente encore $\alpha$.

- $\alpha = 9$

```python
w, b = artificial_neuron(X, y, learning_rate = 9, n_iter = 100)
```

On a toujours une loss qui a une tendance baissière mais les sauts sont plus drastiques. Gain sur l'accuracy minime. On a donc augmenter l'instabilité de notre modèle sans gain substantiel. Ces hyperparamètres sont à éviter : on aurait eu un modèle dans les choux si l'on s'était arrêté à 40 itérations par exemple.

- $\alpha = 0.01$

```python
w, b = artificial_neuron(X, y, learning_rate = 0.01, n_iter = 100)
```

Convergence OK

- $\alpha = 0.005$

```python
w, b = artificial_neuron(X, y, learning_rate = 0.005, n_iter = 100)
```

Le modèle ne converge plus

Conclusion :
- Un learning rate trop élevé risque de rendre l'apprentissage instable
- Un learning rate trop faible empêche la convergence de l'apprentissage

##### Modification du nombre d'itérations

Evidemment, un nombre d'itération trop faible empêche la convergence. On fait un premier test avec un learning rate faible et un nombre d'itérations élevé.

- $\alpha = 0.005$
- $n = 1k$

```python
w, b = artificial_neuron(X, y, learning_rate = 0.005, n_iter = 1000)
```

Le modèle converge mais on a perdu sur l'accuracy => Un learning rate bien choisit peut permettre de réduire drastiquement le nombre de calculs nécessaires à la convergence.

- $\alpha = 0.005$
- $n = 10k$

```python
w, b = artificial_neuron(X, y, learning_rate = 0.005, n_iter = 10000)
```

Avec ce learning rate faible, il aura fallu plus de 2k boucles pour converger contre environ 50 avec $\alpha = 1$ ! C'est peu dérengeant avec un modèle simple comme le perceptron mais ça peut devenir un cauchemar avec des modèles à plusieurs couches !

##### Augmentation du nombre de données

C'est un hyperparamètre un peu caché mais qui a son importance car il peut impliquer des effets de bord sur notre modèle comme on va le voir. On reprend notre modèle "worst case" ($\alpha = 0.5, n = 100$) mais cette fois-ci avec dix fois plus de data

```python
X, y = make_blobs(n_samples = 1000, n_features = 2, centers = 2, random_state = 0)
y = y.reshape((y.shape[0], 1))

plt.scatter(X[:, 0], X[:, 1], c = y, cmap = 'summer')

plt.show()
```

- $\alpha = 0.005$
- $n = 100$

```python
w, b = artificial_neuron(X, y, learning_rate = 0.005, n_iter = 100)
```

Phénomène intéressant, le learning rate est tellement mauvais que même en augmentant le nombre de données, on n'arrive pas à converger en 100 itérations.

Par contre, on remarque que l'accuracy n'a pas atteint de plateau, même au bout de 100 itérations. On essaie avec 200 itérations.

- $\alpha = 0.005$
- $n = 200$

```python
w, b = artificial_neuron(X, y, learning_rate = 0.005, n_iter = 200)
```

On était effectivement plus très loin de la convergence. En environ 120 itérations on dépasse les 0.9 d'accuracy !

On refait le même test mais avec 100k données. Pour voir si l'on converge encore plus vite ou non.

```python
X, y = make_blobs(n_samples = 100000, n_features = 2, centers = 2, random_state = 0)
y = y.reshape((y.shape[0], 1))

plt.scatter(X[:, 0], X[:, 1], c = y, cmap = 'summer')

plt.show()
```

- $\alpha = 0.005$
- $n = 1000$

```python
w, b = artificial_neuron(X, y, learning_rate = 0.005, n_iter = 1000)
```

Malgré un dataset plus imposant, il nous aura fallu beaucoup plus d'itérations pour converger. Cela provient peut-être du fait que la variance du gradient a augmentée et que le learning rate trop lent n'arrive pas à compenser. On le vérifie en augmentant le learning rate.

- $\alpha = 0.01$
- $n = 1000$

```python
w, b = artificial_neuron(X, y, learning_rate = 0.01, n_iter = 1000)
```

Cela signifie qu'augmenter le nombre de données disponibles ne suifft pas pour avoir un modèle performant. Les hyperparamètres jouent un rôle primordiale dans la convergence du modèle
